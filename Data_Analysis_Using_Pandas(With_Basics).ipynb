```python
cc=pd.DataFrame(np.random.rand(3,2),index=['A','B','C'],columns=[1,2])
dd=pd.DataFrame(np.random.rand(3,3),index=['A','B','C'],columns=[3,4,5])
```

```python
cc
```

```python
dd
```

```python
pd.concat([cc,dd],axis=1)      ### This combines the dataframes by row; remember we have to write axis=1 for this; otherwise the default value is axis=0
```

```python
ee=pd.DataFrame(np.random.rand(4,3),index=['D','F','E','G'],columns=[3,4,5])
```

```python
ee
```

```python
pd.concat([dd,ee],axis=0)   ## This is concatenation by Rows with the help of axis=0; Note if we don't write axis=0; the default value taken would be axis=0 itself
```

```python
pd.melt(ee,id_vars=3,value_vars=[4,5],var_name='No',value_name='Yes')       ## This function will convert rows into columns as seen below with the representation of numbers
```
